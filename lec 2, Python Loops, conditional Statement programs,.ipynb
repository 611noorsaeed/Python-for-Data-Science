###  1 Conditional Statement

1.1 if: used to check a single condition and execute code if true.

1.2 if else: used to check two conditions where one is default if the first fails.

1.3 elif: used for checking multiple conditions one after another.


### 2 Loops

2.1 for loop: used to iterate over a sequence like list, string, range.

2.2 while loop: used to repeat code as long as a condition is true.

# 1. Even and Odd Numbers (Using for and if)

```python
for i in range(1,11):
    if i%2 == 0:
        print('Even number : ',i)
    else:
        print("odd number : ",i)
```

# 3.  Factorial of a Number (Using for and if)

```python
num = int(input("Enter a number to find its factorial: "))

if num < 0:
    print("Factorial not defined for negative numbers.")
elif num == 0 or num == 1:
    print("Factorial is 1.")
else:
    fact = 1
    for i in range(1, num + 1):
        fact *= i
    print(f"Factorial of {num} is {fact}")
```
