#  What is a Scatter Plot?
A scatter plot is a type of data visualization that uses dots to represent the values obtained for two different 
variables â€“ one plotted along the x-axis and the other along the y-axis. It is used to observe relationships,
correlations, or patterns between the two variables.


# When to Use a Scatter Plot?

1. You want to visualize the relationship or correlation between two numerical variables.

2. You suspect a trend (positive, negative, or no correlation).

3. You want to identify outliers or clusters in the data.

4. You want to test if a change in one variable affects another.


#  Which Data Goes on the X and Y Axes?

A) X-axis: Typically the independent variable (the one you control or assume causes change).

B) Y-axis: The dependent variable (the one that changes based on the X value).

### For example:

If you're studying how temperature (x) affects ice cream sales (y), then:

X = temperature (independent)

Y = ice cream sales (dependent)

# USAGE

```python
import numpy as np 
import matplotlib.pyplot as plt 


# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.colorbar()
plt.show()
```



# Example 1: Hours Studied vs Exam Score

```python
x = [1, 2, 3, 4, 5]
y = [50, 55, 65, 70, 85]

plt.scatter(x, y, color='blue')
plt.title("Hours Studied vs Exam Score")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

#  Example 2: Car Speed vs Braking Distance

```python
x = [30, 40, 50, 60, 70]  # Car speed in km/h
y = [14, 20, 30, 45, 60]  # Braking distance in meters

plt.scatter(x, y, color='red')
plt.title("Car Speed vs Braking Distance")
plt.xlabel("Car Speed (km/h)")
plt.ylabel("Braking Distance (m)")
plt.grid(True)
plt.show()
```
