# What is a Line Plot?
A line plot is a type of chart that displays information as a series of data points connected by straight line segments. It's typically used to show trends over time or to depict the relationship between two continuous variables.




# When to Use a Line Plot

1 You want to show trends, changes, or progress over a continuous interval (like time, age, temperature).

2 You have ordered categories on the x-axis (e.g., days, years, or experiment steps).

3 You're comparing one or more sets of continuous data.


# Types of Data 


X-axis	Continuous / Sequential	Time (years, months, seconds), Age, Steps in a process

Y-axis	Continuous / Measurable	Height, Temperature, Profit, Score

# Line Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Proper sample data
height = [100, 120, 130, 150, 160, 170, 180, 190, 200]  # in centimeters
age = [3, 4, 7, 10, 15, 20, 30, 40, 50]                # in years

# Line Plot
plt.plot(age, height, label='Height vs Age', color='blue', linestyle='--', linewidth=2, marker='o')
plt.title('Height vs Age')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.legend()
plt.grid(True)
plt.show()
```

# Real Data

```python
import pandas as pd 
df = pd.read_csv("lec 17, dataset.csv")


# Line Plot
plt.plot(df['Age'], df['Fare'], label='Age Vs Fare', color='blue', linestyle='--', linewidth=2, marker='o')
plt.title('Age Vs Fare')
plt.xlabel('Age (years)')
plt.ylabel('Fare (points)')
plt.legend()
plt.grid(True)
plt.show()
```
