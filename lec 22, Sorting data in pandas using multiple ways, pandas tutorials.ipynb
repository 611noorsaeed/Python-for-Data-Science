# Load Dataset

```python
import pandas as pd
df = pd.read_csv("lec 17, dataset.csv")
df.head()
```

# 1 Sorting by a single column (e.g., Age)

```python
df.sort_values(by='Age')
```

```python
df.sort_values(by="Name")
```

# 2 Sorting by multiple columns

```python
df.sort_values(by=['Name','Age'])
```

# 3 Sorting By Index

```python
df.sort_index()
```

# 4 Sorting in descending order

```python
df.sort_values(by='Age',ascending=False)
```

# 5 Sorting and keeping the original index

```python
df.sort_values(by='Age').reset_index(drop=True)
```

# 6 Sorting by a custom order

```python
# Sorting by a custom function (e.g., sorting by name length)
df.reindex(df['Name'].str.len().sort_values().index)
```

# 7 Sorting With Lambda and Ignoring Missing Values

```python
df.sort_values(by='Age',key=lambda x:x.fillna(0))
```
